#### Problem:

- In the first data collection (for both Gaza-Israel & Ukraine-Russia), all fact-checking articles from July, 1st, 2024 to April, 30th, 2025 were considered.
- Some of the articles from July 2024 also contain claims from June 2024, which could have been part of the pre-training data of the LLM (Gemini 2.0 Flash-lite) used in the evaluation, since the knowledge cutoff date of this LLM is June 2024.
- Thus, the claims with date *before* July 2024 will be removed. As a consequence, the claim dates of both the manual and the api part of the dataframes (Gaza-Israel & Ukraine-Russia) are consistent.

```python
# load libraries

import pandas as pd
```

```python
# load manual df 

gaza_israel_manual_df = pd.read_csv("gaza_israel_manual_dataset_010724_300425_old.csv", index_col = 0)
print(len(gaza_israel_manual_df))
print(gaza_israel_manual_df.dtypes)
gaza_israel_manual_df.head()
```

```python
# Transform both the claim and review date columns into datetime format

gaza_israel_manual_df["Claim_Date"] = pd.to_datetime(gaza_israel_manual_df["Claim_Date"], format='%Y-%m-%d')
gaza_israel_manual_df["Review_Date"] = pd.to_datetime(gaza_israel_manual_df["Review_Date"] , format='%Y-%m-%d')

# check the data types
print(gaza_israel_manual_df.dtypes)

gaza_israel_manual_df.head()
```

```python
# 1st: look at all claims with date from 2024 (get overview, sanity check)

gaza_israel_manual_df.loc[(gaza_israel_manual_df["Claim_Date"] < '2025-01-01')]
```

```python
# 2nd: look at all claims with date before July 2024 

df_claims_to_remove = gaza_israel_manual_df.loc[(gaza_israel_manual_df["Claim_Date"] < '2024-07-01')]
print(len(df_claims_to_remove))
df_claims_to_remove
```

- Will lose 4 text-only claims and 3 claims with normal image
- Will not lose any claims with AI-generated or altered images

```python
## Only keep claims with dates from July, 1st, 2024 on and save it as new df

df_claims_to_keep = gaza_israel_manual_df.loc[(gaza_israel_manual_df["Claim_Date"] > '2024-06-30')]
print(len(df_claims_to_keep))

df_claims_to_keep.head(20)

## Reindex the df
df_claims_to_keep = df_claims_to_keep.reset_index(drop=True)
df_claims_to_keep['id'] = range(len(df_claims_to_keep))

df_claims_to_keep.head(20)
```

```python
df_claims_to_keep.to_csv("gaza_israel_manual_dataset_010724_300425.csv")
df_claims_to_keep.head()
```
